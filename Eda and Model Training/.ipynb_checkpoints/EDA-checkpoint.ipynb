```python
sns.countplot(x='Classes', data=df)
plt.title("Target Variable Distribution (Fire vs Not Fire)")

plt.show()
```

```python
df.describe()
```

```python
df.hist(figsize=(20,15), bins=20)
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=df.drop(columns=['Classes']))
plt.xticks(rotation=90)
plt.title("Outlier Visualization using Boxplot")
plt.show()
```

```python
sns.boxplot(x='Classes', y='FWI', data=df)
plt.title("FWI vs Fire Occurrence")
plt.show()
```

```python
sns.boxplot(x='Classes', y='Temperature', data=df)
plt.title("Temperature vs Fire Occurrence")
plt.show()
```

```python
sns.pairplot(
    df[['Temperature', 'RH', 'FWI', 'ISI', 'Classes']],
    hue='Classes'
)
plt.show()
```
